```python
from re import findall
cmpt = 0
for entity, wiki in database.wikipedia['found'].items():
    try:
        title = findall(r'(.*?)\s*\(', wiki.title)[0]
    except IndexError:
        title = wiki.title
    if title != entity:
        cmpt += 1
        print(wiki.title + '/' + entity)
print(cmpt)
```
